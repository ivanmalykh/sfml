#### (6) Создать новый строковый столбец в data frame-е, который:
- принимает значение A, если значение LIMIT_BAL <=10000
- принимает значение B, если значение LIMIT_BAL <=100000 и >10000
- принимает значение C, если значение LIMIT_BAL <=200000 и >100000
- принимает значение D, если значение LIMIT_BAL <=400000 и >200000
- принимает значение E, если значение LIMIT_BAL <=700000 и >400000
- принимает значение F, если значение LIMIT_BAL >700000

```python
%%time

def class_of_litim_bal(val):
    if val <= 10000:
        return 'A'
    elif (val <= 100000) & (val > 10000):
        return 'B'
    elif (val <= 200000) & (val > 100000):
        return 'C'
    elif (val <= 400000) & (val > 200000):
        return 'D'    
    elif (val <= 700000) & (val > 400000):
        return 'E'
    else:
        return 'F'

df.LIMIT_BAL.map(class_of_limit_bal)
#df['LIMIT_BAL_CLASS'] = df.LIMIT_BAL.map(class_of_limit_bal)
```

```python
%%time
A = [[a,b] for a in df.LIMIT_BAL[df.LIMIT_BAL <= 10000].index.tolist() for b in'A']
B = [[a,b] for a in df.LIMIT_BAL[(df.LIMIT_BAL <= 100000) & (df.LIMIT_BAL > 10000)].index.tolist() for b in 'B']
C = [[a,b] for a in df.LIMIT_BAL[(df.LIMIT_BAL <= 200000) & (df.LIMIT_BAL > 100000)].index.tolist() for b in 'C']
D = [[a,b] for a in df.LIMIT_BAL[(df.LIMIT_BAL <= 400000) & (df.LIMIT_BAL > 200000)].index.tolist() for b in 'D']
E = [[a,b] for a in df.LIMIT_BAL[(df.LIMIT_BAL <= 700000) & (df.LIMIT_BAL > 400000)].index.tolist() for b in 'E']
F = [[a,b] for a in df.LIMIT_BAL[(df.LIMIT_BAL > 700000)].index.tolist() for b in 'F']

new_col_limit_bal_class = A+B+C+D+E+F

new_col_limit_bal_class = sorted(new_col_limit_bal_class)
df['LIMIT_BAL_CLASS'] = [x[1] for x in new_col_limit_bal_class]
```
